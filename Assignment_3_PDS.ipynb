```python
# Stage-1 Importing Libraries

import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Stage-2 Loading data file into a Pandas DataFrame
Datafile = pd.read_csv('/Corona_NLP_test.csv')
```

**Solution-1 Converting the text corpus into tokens** **bold text**

```python
# Stage-3 Converting the text corpus into tokens
Corpus = Datafile['OriginalTweet'].apply(lambda x: word_tokenize(str(x)))
```

**Solution-2 Performing stop word removal** 

```python
# Stage-4 Performing stop word removal
StopWords = set(stopwords.words('english'))
Corpus = Corpus.apply(lambda x: [word for word in x if word.lower() not in StopWords])
```

**Solution-3 Counting Word frequencies**

```python
# Step-5 Counting Word frequencies
FlatCorpus = [word for tweet in Corpus for word in tweet]
WordFreq = FreqDist(FlatCorpus)
```

**Solution-4 Creating word clouds**

```python
# Step-6 Creating word clouds
WordCloud = WordCloud(width=800, height=400, background_color='white', max_words=50).generate_from_frequencies(WordFreq)
# Step-7 Ploting the word cloud
plt.figure(figsize=(12, 10))
plt.imshow(WordCloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
